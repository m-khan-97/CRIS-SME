# CRIS-SME Live Report Figures

This notebook converts the current `cris_sme_report.json` output into lightweight SVG figures for demos, papers, and case-study documentation.

```python
import sys
from pathlib import Path
import json

repo_root = Path('..').resolve()
src_path = repo_root / 'src'
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from cris_sme.reporting.figure_export import write_report_figures

report_path = repo_root / 'outputs' / 'reports' / 'cris_sme_report.json'
output_dir = repo_root / 'outputs' / 'figures'

report = json.loads(report_path.read_text(encoding='utf-8'))
figure_paths = write_report_figures(report, output_dir)
figure_paths
```

```python
for label, path in figure_paths.items():
    print(f'{label}: {path.resolve()}')
```

## Notes

- The figures are generated as SVG so they can be embedded directly into markdown documentation and presentation slides.
- The chart logic lives in `src/cris_sme/reporting/figure_export.py` so it stays testable and reusable outside the notebook.
